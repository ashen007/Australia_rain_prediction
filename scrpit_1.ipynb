## import packages and data

```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('../../case data/australia rain/weatherAUS.csv')
data.head(6)
```

```python
train,test = train_test_split(data,test_size=0.2,random_state=42)
train.shape, test.shape
```

```python
sample = train[['MaxTemp','WindGustSpeed','RainTomorrow']].dropna()
sample
```

```python
sample.info()
```

```python
plt.figure(figsize=[12,6],dpi=300)
sns.scatterplot(x='MaxTemp',y='WindGustSpeed',data=sample[sample.RainTomorrow=='Yes'],)
sns.scatterplot(x='MaxTemp',y='WindGustSpeed',data=sample[sample.RainTomorrow=='No'])
plt.show()
```
